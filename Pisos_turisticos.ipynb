```python
import geopandas as gpd
import matplotlib.pyplot as plt
import os

# Ruta a tu shapefile
shapefile_path = "data/map"  # Cambia por tu ruta real

# Cargar shapefile
gdf = gpd.read_file(shapefile_path)

# Carpeta para guardar SVGs individuales
output_dir = "salida_euskadi"
os.makedirs(output_dir, exist_ok=True)

# Guardar cada polígono en un SVG individual
for i, row in gdf.iterrows():
    fig, ax = plt.subplots(figsize=(6, 6))
    gdf.iloc[[i]].plot(ax=ax, edgecolor='black', facecolor='none')  # Solo contorno
    ax.axis('off')
    plt.tight_layout()
    filename = os.path.join(output_dir, f"poligono_{i}.svg")
    plt.savefig(filename)
    plt.close()
    print(f"Guardado: {filename}")

# Guardar todos los polígonos juntos en un solo SVG
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, edgecolor='black', facecolor='none')
ax.axis('off')
plt.tight_layout()
all_polygons_file = "salida_euskadi/todos_los_poligonos.svg"
plt.savefig(all_polygons_file)
plt.close()
print(f"Guardado todos los polígonos juntos: {all_polygons_file}")
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import os

gdf = gpd.read_file("data/map")
df = pd.read_csv("data/comarcas.csv")

# Normalizar nombres para el merge
gdf["COMARCA"] = gdf["COMARCA"].str.strip().str.upper()
df["COMARCA"] = df["COMARCA"].str.strip().str.upper()

merged = gdf.merge(df, on="COMARCA")

anios = [str(a) for a in range(2006, 2024)]

os.makedirs("salida_mapas", exist_ok=True)

for anio in anios:
    fig, ax = plt.subplots(figsize=(10, 8))
    merged.plot(
        column=anio,
        cmap='YlOrRd',
        linewidth=0.5,
        ax=ax,
        edgecolor='0.8',
        legend=True,
        legend_kwds={'shrink': 0.7}  # Reducir tamaño barra color
    )

    # Añadir etiquetas con los nombres de las comarcas
    for idx, row in merged.iterrows():
        # Obtener coordenadas del centroide
        x, y = row['geometry'].centroid.x, row['geometry'].centroid.y
        ax.text(x, y, row['COMARCA'], fontsize=6, color='#444444', ha='center', va='center')

    ax.set_title(f"Licencias por comarca - {anio}", fontsize=14)
    ax.axis('off')
    plt.tight_layout()

    # Ajustar tamaño letra leyenda (colorbar)
    cbar = ax.get_figure().axes[-1]  # La barra de color suele ser el último axis
    cbar.yaxis.label.set_size(8)    # Tamaño letra de la etiqueta
    for tick in cbar.get_yticklabels():
        tick.set_fontsize(7)         # Tamaño letra de los ticks

    plt.savefig(f"output_mapas_comarcas/mapa_{anio}.png", dpi=150)
    plt.close()
```
